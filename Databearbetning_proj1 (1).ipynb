##Del 1: läsa in data



####- Skapa tre variabler med passande kolumner och läs in användardata, ratingdata ochfilmdata i variablerna.

```python
import pandas as pd

#dictionary med urlarna till data, för ez access
urls = {"ratings": "https://files.grouplens.org/datasets/movielens/ml-100k/u.data",
        "users" : "https://files.grouplens.org/datasets/movielens/ml-100k/u.user",
        "movies": "https://files.grouplens.org/datasets/movielens/ml-100k/u.item"}
#Eftersom data inte innehåller några headers sätter vi in egna, also from dictionary for ez access       
dfHeaders = {"ratings": ["user id" , "item id" , "rating" , "timestamp"],
             "users" : ["id", "Age", "Gender", "Occupation", "zipcode"],
             "movies": ["movie id" , "movie title" , "release date" , "video release date", "IMDb URL" , "unknown" , "Action" , "Adventure" ,
                        "Animation", "Children's" , "Comedy" , "Crime" , "Documentary" , "Drama", "Fantasy", "Film-Noir" , "Horror" ,
                        "Musical" , "Mystery" , "Romance","Sci-Fi", "Thriller","War" , "Western"]}
#här läser vi in datat till variablerna
userDf = pd.read_csv(urls["users"], sep="|", header=None, names=dfHeaders["users"])
ratingsDf = pd.read_csv(urls["ratings"], header=None, names=dfHeaders["ratings"],delim_whitespace=True)
#jag mindes inte vilken encoding du sade att man skulle använda, och hade inte tillgång till videona från lektion. Prövade mig fram, men den där "cp437" verkar fungera 
moviesDf = pd.read_csv(urls["movies"],sep="|", header=None, names=dfHeaders["movies"], encoding="cp437")
```

##Del 2: Välj Data

####- Sortera ratings enligt film. (Vi vill alltså se alla ratings för en film under varann)

```python
#här sorterar vi ratings enligt film ID
sortedratingsDf = ratingsDf.sort_values("item id", ascending=True)
sortedratingsDf
```

##Del 3: Filtrera Data

####- Visa endast användare med ett visst yrke.
####- Visa endast användare som är över 40 och män.


```python
#här tar vi från Users datan alla engineers över 40 år gamla, på samma gång ser vi att det inte finns några kvinnor med de kriterierna så det behöver inte filtreras
engineerDf = userDf.loc[(userDf["Occupation"] == "engineer") & (userDf["Age"] >= 40)]
engineerDf
```

##Del 4: Utforska data

####- Visa medelåldern av användarna som är författare.

####- Använd den tidigare ratings per film sorteringen för att räkna medelrating per film.

```python
#här gruperar vi alla ratings med samma "item id" till en lista och sedan lagar vi en ny dataframe med kolumnerna "item id" som är filmens ID
#och "rating" som är en lista med alla ratings för den filmen
grouped_sortedratingsDf = sortedratingsDf.groupby("item id")
grouped_ratings = grouped_sortedratingsDf["rating"].apply(list)
grouped_ratings = grouped_ratings.reset_index()
```

```python
grouped_ratings.head()

```

```python
from statistics import mean

#här tar vi ut en list of lists som innehåller alla ratings
#och kalkylerar ut medeltalet av varje list och sätter svaret i en ny lista
ratinglist = grouped_ratings["rating"].tolist()
meanratinglist = []
for x in range(len(ratinglist)):
  meanratinglist.append(mean(ratinglist[x]))
#sedan lägger vi listan med medeltalen tillbaka i vår gruperade dataframe
grouped_ratings["average_rating"] = meanratinglist
```

```python
grouped_ratings
```

####- Hitta top 10 filmerna bland de som har fler än 40 ratings.

```python
#här lagar jag en till collumn med data för hur många ratings varje film har
grouped_ratings["number of ratings"] = grouped_ratings["rating"].apply(lambda x:len(x))  
#och tar ut en dataframe var "number of ratings" är större än 40
grouped_morethan40ratings = grouped_ratings.loc[grouped_ratings['number of ratings'] >= 40]
```

```python
#sorterar den enligt medeltalet av rating
grouped_morethan40ratings_sorted = grouped_morethan40ratings.sort_values("average_rating", ascending=False)
#printar top10
grouped_morethan40ratings_sorted.head(10)
```

Här söker vi fram de populäraste arbeten bland män i vårt dataset.

```python
maleUserDf = userDf.loc[(userDf["Gender"] == "M")]

#grupera ID enligt "Occupation"
grouped_maleUsersDf = maleUserDf.groupby("Occupation")
grouped_occupation = grouped_maleUsersDf["id"].apply(list)
grouped_occupation = grouped_occupation.reset_index()
#laga en column med mängden arbetare i varje kategori
grouped_occupation["number of occupation"]  = grouped_occupation["id"].apply(lambda x:len(x))  
#sortera minskande "number of occupation"
grouped_occupation = grouped_occupation.sort_values("number of occupation", ascending=False)
#populäraste arbeten bland män
grouped_occupation
```

Samma som ovan bara att för kvinnor

```python
femaleUserDf = userDf.loc[(userDf["Gender"] == "F")]

#grupera ID enligt "Occupation"
grouped_femaleUsersDf = femaleUserDf.groupby("Occupation")
grouped_occupation_fe = grouped_femaleUsersDf["id"].apply(list)
grouped_occupation_fe = grouped_occupation_fe.reset_index()
#laga en column med mängden arbetare i varje kategori
grouped_occupation_fe["number of occupation"]  = grouped_occupation_fe["id"].apply(lambda x:len(x))  
#sortera minskande "number of occupation"
grouped_occupation_fe = grouped_occupation_fe.sort_values("number of occupation", ascending=False)
#populäraste arbeten bland kvinnor
grouped_occupation_fe
```
